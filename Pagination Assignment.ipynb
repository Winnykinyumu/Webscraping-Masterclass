```python
#Importing the necessary libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#Inspecting the website
url = 'https://books.toscrape.com/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify)
```

## **Extracting Title | Price | Star Rating |Availability**

```python
#Initiating an empty variable to store the output
books=[]

#Looping through the pages
for i in range(1,6):
  url3 = f"https://books.toscrape.com/catalogue/page-{i}.html"
  response3 = requests.get(url3)
  response3 = response3.content
  soup3 = BeautifulSoup(response3, 'html.parser')
  ol = soup3.find('ol')
  articles = ol.find_all('article', class_='product_pod')
  for article in articles:
    image = article.find('img')
    title = image.attrs['alt']
    startag = article.find('p')
    star = startag['class'][1]
    price = article.find('p', class_='price_color').text
    price = float(price[1:])
    availability = article.find('p', class_="instock availability").text.strip()

#Storing the outputs in the books variable
    books.append([title, star, price, availability])

#Printing the output
    print(books)
```

```python
# Converting the output into a pandas Dataframe

df = pd.DataFrame(books, columns=['Title', 'Star Rating', 'Price','Availability'])
df
```

```python
#Saving the output into a CSV file
df.to_csv('my_books.csv')
```
