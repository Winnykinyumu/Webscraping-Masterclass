```python
url2 = 'https://books.toscrape.com/catalogue/page-1.html'
response2 = requests.get(url2)
soup2 = BeautifulSoup(response2.text, 'html.parser')

books = soup2.find_all('article', class_='product_pod')

for book in books:
  title = book.h3.a['title']
  price = book.find('p', class_="price_color").text
  availability = book.find('p', class_="instock availability").text

print(f'Title: {title} | Price: {price} | Availability: {availability}')
```

# SCRAP MULTIPLE PAGES

```python
import pandas as pd
```

```python
books = []

for i in range(1,6):
  url3 = f"https://books.toscrape.com/catalogue/page-{i}.html"
  response3 = requests.get(url3)
  response3 = response3.content
  soup3 = BeautifulSoup(response3, 'html.parser')
  ol = soup3.find('ol')
  articles = ol.find_all('article', class_='product_pod')
  for article in articles:
    image = article.find('img')
    title = image.attrs['alt']
    startag = article.find('p')
    star = startag['class'][1]
    price = article.find('p', class_='price_color').text
    price = float(price[1:])

    books.append([title, star, price])
```

```python
print(books)
```

## Converting output into a pandas DataFrame

```python
df = pd.DataFrame(books, columns=['Title', 'Star Rating', 'Price'])
```

```python
df.head(10)
```

# Convert our output to a csv file

```python
df.to_csv('Books.csv')
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import service
from  selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(options=chrome_options)
```

# Define website url

```python
website = 'https://www.adamchoi.co.uk/teamgoals/detailed'
driver.get(website)
```

```python
from selenium.webdriver.common.by import By

#locate elements tha we need
all_matches_button = driver.find_element(By.XPATH, '//label[@analytics-event="All matches"]')

all_matches_button.click()
```

```python
date = []
home_team = []
score = []
away_team =[]


matches = driver.find_elements(By.TAG_NAME, 'tr')

for mattch in matches:
  try:
    date.append(mattch.find_element(By.XPATH, './/td[1]').text)
    home = mattch.find_element(By.XPATH, './/td[2]').text
    home_team.append(home)
    score.append(mattch.find_element(By.XPATH, './/td[3]').text)
    away_team.append(mattch.find_element(By.XPATH, './/td[4]').text)

    print(f'{home} vs {away_team[-1]} | score: {score[-1]} | date: {date[-1]}')

  except:
    # no error
    continue
```

```python
import pandas as pd
df5 = pd.DataFrame({'date': date, 'home_team': home_team, 'score': score, 'away_team': away_team})
df5.head()
```

```python
df5.to_csv('results.csv')
```

```python
from selenium.webdriver.support.ui import Select
dropdown_element = driver.find_element(By.ID, 'country')

dropdown = Select(dropdown_element)

dropdown.select_by_visible_text('Spain')
```

```python
matches = driver.find_elements(By.TAG_NAME, 'tr')

for matchh in matches:
    try:
        # Correct usage with By.XPATH and relative XPath
        date.append(matchh.find_element(By.XPATH, './/td[1]').text)
        home = matchh.find_element(By.XPATH, './/td[2]').text
        home_team.append(home)
        score.append(matchh.find_element(By.XPATH, './/td[3]').text)
        away_team.append(matchh.find_element(By.XPATH, './/td[4]').text)

        # Optional: print results
        print(f'{home} vs {away_team[-1]} | Score: {score[-1]} | Date: {date[-1]}')

    except:
        # Skip rows that don't contain expected <td> elements
        continue
```
